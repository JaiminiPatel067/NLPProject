```python
import os
import PyPDF2
from transformers import pipeline
import torch  # Import torch to check for GPU availability

# Function to extract text from a PDF
def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page in range(len(reader.pages)):
            page_text = reader.pages[page].extract_text()
            if page_text:
                text += page_text
    return text.strip()

# Check if GPU is available and set device accordingly
device = 0 if torch.cuda.is_available() else -1  # Use GPU if available, otherwise CPU

# Initialize the summarizer pipeline (using CPU if no GPU available)
summarizer = pipeline("summarization", model="facebook/bart-large-cnn", tokenizer="facebook/bart-large-cnn", device=device)

# Path to the folder containing PDFs (use '/content/' as your base directory in Colab)
folder_path = '/content/'

# List all the PDF files in the folder (no need for subfolder since they're directly in /content/)
pdf_files = [f for f in os.listdir(folder_path) if f.startswith('shortlist_') and f.endswith('.pdf')]

# Loop through each PDF file
for pdf_file in pdf_files:
    pdf_path = os.path.join(folder_path, pdf_file)

    # Extract text from the PDF
    pdf_text = extract_text_from_pdf(pdf_path)

    # Check if extracted text is too short
    if len(pdf_text) < 50:
        print(f"Skipping {pdf_file} as it's too short for summarization.")
        continue

    # Check the length of the extracted text and adjust the max_length for summarization
    input_length = len(pdf_text.split())
    max_len = min(input_length // 2, 150)  # Adjust to half of input text length (max 150 tokens)

    # Ensure the text length is within a safe range for summarization (e.g., no more than 1024 tokens for BART)
    if input_length > 1024:
        pdf_text = pdf_text[:1024]  # Truncate to the first 1024 tokens

    # Perform summarization
    try:
        summary_output = summarizer(pdf_text, max_length=max_len, min_length=10, do_sample=False)
        # Print the summary for each file
        print(f"Summary for {pdf_file}:")
        print(summary_output[0]['summary_text'])
    except Exception as e:
        print(f"Error summarizing {pdf_file}: {e}")

    print("\n" + "-"*50 + "\n")
```

```python
from sklearn.metrics import f1_score
from rouge_score import rouge_scorer
import nltk
from nltk.translate.bleu_score import sentence_bleu

# Function to compute ROUGE scores
def compute_rouge(reference, summary):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = scorer.score(reference, summary)
    return scores

# Function to compute BLEU score
def compute_bleu(reference, summary):
    reference_tokens = reference.split()
    summary_tokens = summary.split()
    return sentence_bleu([reference_tokens], summary_tokens)

# Function to compute F1 score (if you have binary classification)
def compute_f1(reference, summary):
    # Example: You can compare word presence in both reference and summary for a simple F1 computation
    reference_tokens = set(reference.split())
    summary_tokens = set(summary.split())
    common_tokens = reference_tokens.intersection(summary_tokens)

    if len(common_tokens) == 0:
        return 0.0

    precision = len(common_tokens) / len(summary_tokens)
    recall = len(common_tokens) / len(reference_tokens)

    if precision + recall == 0:
        return 0.0

    f1 = 2 * (precision * recall) / (precision + recall)
    return f1

# Example of generating summaries and comparing them (fill this with actual generated and reference summaries)
generated_summaries = ["This is a generated summary example."]
reference_summaries = ["This is a reference summary example."]

# Assuming the reference summaries are your ground truth
for ref_summary, gen_summary in zip(reference_summaries, generated_summaries):
    # Compute ROUGE scores
    rouge_scores = compute_rouge(ref_summary, gen_summary)
    print(f"ROUGE scores for generated summary: {rouge_scores}")

    # Compute BLEU score
    bleu_score = compute_bleu(ref_summary, gen_summary)
    print(f"BLEU score for generated summary: {bleu_score}")

    # Compute F1 score (here, using a simple approach of token comparison)
    f1 = compute_f1(ref_summary, gen_summary)
    print(f"F1 score for generated summary: {f1}")

    print("-" * 50)
```

```python
import os
import nltk
from transformers import T5Tokenizer, T5ForConditionalGeneration
from rouge_score import rouge_scorer
from nltk.translate.bleu_score import sentence_bleu
from sklearn.metrics import f1_score
import PyPDF2

# Function to extract text from a PDF
def extract_text_from_pdf(pdf_path):
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page in range(len(reader.pages)):
            text += reader.pages[page].extract_text()
    return text

# Initialize the T5 model and tokenizer
model_name = "t5-large"  # You can use t5-small or t5-base for faster performance
model = T5ForConditionalGeneration.from_pretrained(model_name)
tokenizer = T5Tokenizer.from_pretrained(model_name)

# Function to generate summary using T5
def generate_summary_t5(text, max_length=150):
    inputs = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = model.generate(inputs, max_length=max_length, min_length=30, length_penalty=2.0, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

# Function to compute ROUGE scores
def compute_rouge_scores(reference, generated):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = scorer.score(reference, generated)
    return scores

# Function to compute BLEU score
def compute_bleu_score(reference, generated):
    reference = reference.split()  # Tokenize reference
    generated = generated.split()   # Tokenize generated summary
    return sentence_bleu([reference], generated)

# Function to compute F1 score (adjusted for token set comparison)
def compute_f1_score(reference, generated):
    reference_tokens = set(reference.split())  # Convert to set of tokens for reference
    generated_tokens = set(generated.split())  # Convert to set of tokens for generated summary

    # Calculate the intersection (common tokens) and F1 score based on that
    common_tokens = reference_tokens.intersection(generated_tokens)
    if len(reference_tokens) == 0:
        return 0.0  # If reference summary is empty, return 0 F1 score

    precision = len(common_tokens) / len(generated_tokens) if len(generated_tokens) > 0 else 0
    recall = len(common_tokens) / len(reference_tokens)
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    return f1

# Path to the content folder where your PDFs are located
folder_path = '/content/'  # Your Google Colab content folder
pdf_directory = folder_path  # PDFs are directly in the content folder

# List all the PDF files in the content folder
pdf_files = [f for f in os.listdir(pdf_directory) if f.endswith('.pdf')]

# Loop through each PDF file and summarize, then evaluate
for pdf_file in pdf_files:
    pdf_path = os.path.join(pdf_directory, pdf_file)

    # Extract text from the PDF
    pdf_text = extract_text_from_pdf(pdf_path)

    # Generate summary using T5
    generated_summary = generate_summary_t5(pdf_text)

    # Assume you have the reference summary (the correct summary for comparison)
    # Here we are using a dummy reference summary for demonstration purposes
    # Replace this with actual reference text as needed
    reference_summary = "Your reference summary text here."

    # Compute ROUGE scores
    rouge_scores = compute_rouge_scores(reference_summary, generated_summary)
    print(f"ROUGE scores for {pdf_file}:")
    print(rouge_scores)

    # Compute BLEU score
    bleu_score = compute_bleu_score(reference_summary, generated_summary)
    print(f"BLEU score for {pdf_file}: {bleu_score}")

    # Compute F1 score
    f1 = compute_f1_score(reference_summary, generated_summary)
    print(f"F1 score for {pdf_file}: {f1}")

    print("\n" + "-"*50 + "\n")
```
